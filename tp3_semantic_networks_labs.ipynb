```python

# Semantics Network Labs
# TP : Réseaux sémantiques
# Nom : HmamedABdelmounaim
# Date : 11-11-2024


import networkx as nx
import matplotlib.pyplot as plt

# Création du graphe
G = nx.DiGraph()  # Utilisation d'un graphe orienté pour préciser les relations directionnelles

# Ajout des nœuds
concepts = ["chien", "animal", "viande", "nourriture", "chat", "être vivant", "respirer"]
G.add_nodes_from(concepts)

# Ajout des arêtes avec les relations
relations = [
    ("chien", "animal", "est un"),
    ("chien", "viande", "mange"),
    ("viande", "nourriture", "est un"),
    ("chat", "animal", "est un"),
    ("animal", "être vivant", "est un"),
    ("être vivant", "respirer", "a besoin de")
]

for start, end, relation in relations:
    G.add_edge(start, end, label=relation)
# Dessiner le réseau
pos = nx.spring_layout(G)  # Génère une disposition pour le graphe

# Dessiner les nœuds et les arêtes
nx.draw(G, pos, with_labels=True, node_size=2000, node_color="skyblue", font_size=10, font_weight="bold", arrows=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, 'label'), font_color="red", font_size=8)

# Afficher le graphe
plt.title("Réseau Sémantique de Concepts et Relations")
plt.show()
```

```python
# Afficher les relations associées au concept "chien"
print("Relations associées au concept 'chien' :")
for neighbor in G.neighbors("chien"):
    relation = G["chien"][neighbor]["label"]
    print(f"chien -> {neighbor} ({relation})")

# Vérifier la relation entre "chien" et "animal"
if G.has_edge("chien", "animal"):
    print("\nRelation entre 'chien' et 'animal' :")
    print(f"chien -> animal ({G['chien']['animal']['label']})")
else:
    print("\nAucune relation directe entre 'chien' et 'animal'.")
```
